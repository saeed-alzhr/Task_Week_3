## Feature Selection
Select relevant features for clustering. Explain your choice of features.


```python
#iam only intrested in nurical values
df.drop(columns=df.select_dtypes(include=['object']).columns,inplace=True)
df.drop("Year",axis=1,inplace=True)
df.drop("Urban_or_Rural_Area",axis=1,inplace=True)
```

```python
df.isna().sum()
```

```python
knn=KNNImputer(n_neighbors=5)
df[["Accident_Severity"]]=knn.fit_transform(df[['Accident_Severity']])
df[["Number_of_Casualties"]]=knn.fit_transform(df[['Number_of_Casualties']])
```

```python
df.isna().sum()
```

## Data Visualization
Visualize the data using appropriate plots to gain insights into the dataset. Using the following:
- Scatter plot of accidents based on Longitude and Latitude.

```python
#plot the longitude and the latitude
plt.figurefigsize=(10, 8)
sns.scatterplot(x='Longitude', y='Latitude', data=df)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python
#Scale the data before the PCA
scaler=StandardScaler()
df=scaler.fit_transform(df)
df=pd.DataFrame(df)
```

```python
#run the pca and the 0,9 is to keep 90% of the knowldge from the data
pca=PCA(n_components=0.9)
df=pca.fit_transform(df)
```

## Clustering
Apply K-Means clustering. Determine the optimal number of clusters and justify your choice.
* Find the `n_clusters` parameter using the elbow method.
* Train the model.

```python
# the number of cluster a tried more the one and saw that four is the best to go with
km = KMeans(n_clusters = 2)
km.fit(df)
cluster_labels = km.fit_predict(df)
```

## Plot the data points with their predicted cluster center

```python
plt.figure(figsize=(10, 8))
plt.scatter(df[:, 0], df[:, 1], c=cluster_labels, cmap='viridis', s=100, alpha=0.7)
plt.scatter(km.cluster_centers_[:, 0], km.cluster_centers_[:, 1], s=300, c='red', marker='X', label='Centroids')
plt.title('Clusters Visualized with PCA')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.legend()
plt.grid(True)
plt.show()
```

## Evaluation
Evaluate the clustering result using appropriate metrics.


```python

from sklearn.metrics import silhouette_score

silhouette_score(df, km.labels_)
```
